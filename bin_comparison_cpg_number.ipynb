```r
data <- read.csv("chr19.csv", col.names = c("bin_id","chr","start","end","input_label","cpg_number","class_label","methylation","read_number","cpg_pattern","A","B","A_norm",
                                           "B_norm","delta","A_sum","B_sum","p_val","p_adj"))

data
```

```r
binID <- data$bin_id
meth <- data$methylation
aNorm <- data$A_norm
bNorm <- data$B_norm
cpg.number <- data$cpg_number
cata <- c()

df <- data.frame(binID, meth, aNorm, bNorm, cpg.number)

df
```

```r
#large.cpg <- df[df$cpg.number > 4, ]

# large.cpg <- large.cpg[order(large.cpg$binID, large.cpg$meth),]

large.cpg <- df[order(df$binID, df$meth),]

temp <- large.cpg$cpg.number
result <- c()

for (i in temp){
    if (! i %in% result){
        result <- append(result, i)
    }
}

result <- sort(result)

result

# for (num in result){
#     pdf(paste("hist of ", num, ".pdf"))
#     tempdf <- df[df$cpg.number == num,]
#     hist(tempdf$meth, main = paste("hist", num))
#     dev.off()
# }

lowStandard <- c()
highStandard <- c()

for (num in result){
    cur.meth <- df[df$cpg.number == num,]$meth
    lowStandard <- append(lowStandard, quantile(cur.meth, probs = c(0.33))[[1]])
    highStandard <- append(highStandard, quantile(cur.meth, probs = c(0.67))[[1]])
}

lowStandard
highStandard

for (i in 1:length(result)){
    if (lowStandard[i] == 0 && highStandard[i] == 0){
        lowStandard[i] <- 32
        highStandard[i] <- 68
    }
    else if (lowStandard[i] == highStandard[i]){
        lowStandard[i] <- floor(lowStandard[i] - (100.0 / (result[i] * 3))) - 1
        highStandard[i] <- ceiling(highStandard[i] + (100.0 / (result[i] * 3))) + 1
    } else if (lowStandard[i] == 0){
        lowStandard[i] <- lowStandard[i] + (100.0 / (result[i] * 3))
    } else{
        lowStandard[i] <- floor(lowStandard[i]) - 1
        highStandard[i] <- ceiling(highStandard[i]) + 1
    }
}

lowStandard
highStandard
```

```r
# Per Bin
# Standard Settings
```

```r
for (i in 1:nrow(df)){
  if (df[i,]$meth < (lowStandard[match(df[i,]$cpg.number, result)])) {
    cata <- append(cata, 1)
  } else if(df[i,]$meth < highStandard[match(df[i,]$cpg.number, result)]){
    cata <- append(cata, 2)
  } else {
    cata <- append(cata, 3)
  }
}


df <- cbind(df, cata)
df
```

```r
df$cata <- factor(df$cata, levels = c(3, 2, 1), labels = c("high", "medium", "low"))
```

```r
df <- df[order(df$binID, df$cata),]
df
```
